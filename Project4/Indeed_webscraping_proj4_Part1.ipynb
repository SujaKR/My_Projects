### Scraping job details from multiple pages

```python
#pages=list(range(0,5000,10))
job_title = []
company_list=[]
locations_list=[]
summary_list=[]
salary_list=[]

for page in tqdm(range(1,201)):
        
    # Multiple by 10 as the numbers follow number of jobs listed per page
    page = (page-1) * 10 
    source= requests.get('https://au.indeed.com/jobs?q=data+scientist&l=Australia&start='+str(page)).text
    soup= BeautifulSoup(source,'lxml')
    
    for jobs in soup.find_all(class_='result'):
        
         
        try:
            titles = jobs.find('div',class_='title').text.strip()
            
        except:
            titles= None
            
        #print('Job title:',titles)
        job_title.append(titles)

        try:
            companies= jobs.find('span',class_='company').text.strip()
        except:
            companies=None
        #print('Company:',companies)
        company_list.append(companies)

        try:
            locations= jobs.find(class_='location').text.strip()
            #locations= jobs.find('span',class_='location').text.strip()
        except :
            locations=None
        
        #print('Locations:',locations)
        locations_list.append(locations)
        

        try:
            salary=jobs.find('span',class_='no-wrap').text.strip()

        except:
            salary= None
        #print('Salary:',salary)
        salary_list.append(salary)

        try:
            summary=jobs.find('div',class_='summary').text.strip()
        except:
            summary=None

        #print('Summary:', summary) 
        summary_list.append(summary)

        #print('-----------------------')
        time.sleep(.5)
        
```

### 5. Create a dataframe

```python
data_job={'Company':company_list,'Location' :locations_list,'Salary':salary_list,'Job Title':job_title,'Summary':summary_list}
```
